```python
import pandas as pd
import seaborn as sn
from scipy import stats
```

```python
temp_meal_data=pd.read_csv('temp_meal_data.csv', index_col=0)
```

```python
temp_meal_data
```

```python
breakfast = temp_meal_data[temp_meal_data['meal'] =='Breakfast']
```

```python
breakfast
```

```python
breakfast.shape
```

```python
dinner = temp_meal_data[temp_meal_data['meal'] =='Dinner']
```

```python
dinner
```

```python
dinner.shape
```

```python
breakfast_sample = breakfast.sample(n=20, replace = True, random_state=0)
```

```python
breakfast_sample
```

```python
dinner_sample = dinner.sample(n=20, replace = True, random_state=0)
```

```python
dinner_sample
```

Null Hypothesis:
There is no difference in temperature (in degrees fahrenheit) on average between students who prefer Breakfast and students who prefer Dinner. In other words, any difference in temperature between these two groups observed in the samples is due to chance.

Alternative Hypothesis:
There is a difference on in temperature (in degrees fahrenheit) on average between students who prefer Breakfast and students who prefer Dinner. In other words, any difference in temperature between these two groups observed in the samples is not due to chance.

```python

#observe diff between the climate of the two groups (breakfast and dinner by using mean)
diff = abs(breakfast_sample['temperature_F'].mean() - dinner_sample['temperature_F'].mean())
```

```python
diff
```

```python
#t test used for small sample size, set equal-var to false
test_results= stats.ttest_ind(a = breakfast_sample['temperature_F'], b = dinner_sample['temperature_F'], equal_var = False)
```

```python
test_results
```

```python
sig_level = 0.01
```

```python
#to get pvalue
p_value=test_results.pvalue
```

```python
p_value
```

```python
#convert to percentage
p_value * 100
```

```python
if p_value < sig_level:
    print(diff, p_value, "Reject")
else:
    print(diff, p_value, "fail to reject")
```
